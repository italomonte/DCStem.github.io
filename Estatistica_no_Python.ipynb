**Estatística Descritiva**

A Estatística Descritiva é um ramo da estatística que se concentra na descrição, organização e sumarização de dados. Seu principal objetivo é fornecer uma compreensão clara e concisa dos padrões, tendências e características presentes em um conjunto de dados. Ao invés de fazer inferências sobre uma população maior com base em amostras, a estatística descritiva se preocupa em extrair informações úteis e insights diretamente do conjunto de dados disponível.

**Objetivos**

**Resumo dos Dados:** A estatística descritiva ajuda a resumir informações complexas em formas mais compreensíveis. Ela apresenta valores-chave, tendências centrais e variações em um formato que pode ser facilmente comunicado e interpretado.

**Identificação de Padrões:** Por meio de gráficos, tabelas e medidas numéricas, a estatística descritiva ajuda a identificar padrões, assimetrias e características notáveis nos dados. Isso pode ser útil para identificar anomalias, outliers e clusters.

**Tomada de Decisões:** A análise descritiva ajuda a embasar tomadas de decisões informadas. Quando os dados são claramente apresentados, os tomadores de decisão podem entender melhor o contexto e fazer escolhas mais acertadas.


**Componentes da Estatística Descritiva**

**1 Medidas de Tendência Central:** Essas medidas indicam onde os valores do conjunto de dados tendem a se concentrar. As principais medidas são a média (valor médio), a mediana (valor central quando os dados estão ordenados) e a moda (valor mais frequente).

**1.1 Média:** A média é a soma de todos os valores dividida pelo número total de valores. É representada por:

$\text{Média} = \frac{\sum_{i=1}^{n} x_i}{n}$

**1.2 Mediana:** A mediana é o valor que separa o conjunto de dados em duas partes iguais, onde metade dos valores está abaixo dela e metade está acima dela. É especialmente útil quando há valores discrepantes (outliers) no conjunto de dados.

**1.3 Moda:** A moda é o valor que aparece com maior frequência no conjunto de dados. Pode haver uma ou mais modas em um conjunto de dados.



```python
import numpy as np
import matplotlib.pyplot as plt

# Exemplo de dados
dados = np.array([12, 15, 18, 21, 24, 15, 18, 21, 18, 21, 24, 27, 21, 24, 24])

# Medidas de Tendência Central
print("Média =",round(np.mean(dados),2))
print("Mediana =",round(np.median(dados),2))
print("Moda =",round(np.argmax(np.bincount(dados)),2))  # Usando a função bincount para encontrar a moda
```



**2 Medidas de Dispersão ou Variabilidade:** Essas medidas quantificam a dispersão ou espalhamento dos valores em relação à tendência central. Incluem o desvio padrão (medida de dispersão em torno da média) e a variância (medida de quanto os valores diferem da média).

**2.1 Variância:** A variância é o quadrado do desvio padrão. Ela também mede a dispersão dos valores em relação à média:

$\text{Variância} = \frac{\sum_{i=1}^{n} (x_i - \text{Média})^2}{n}$

**2.2 Desvio Padrão:** O desvio padrão mede a dispersão dos valores em relação à média. Quanto maior o desvio padrão, maior a variabilidade dos dados:

$\text{Desvio Padrão} = \sqrt{\frac{\sum_{i=1}^{n} (x_i - \text{Média})^2}{n}}$

**2.3 O coeficiente de variação (CV):** é uma medida estatística que expressa a variação relativa em relação à média de um conjunto de dados. Ele é utilizado para comparar a dispersão ou variabilidade de diferentes conjuntos de dados, independentemente de suas unidades de medida:

$\text{Coeficiente de Variação} = \frac{\text{Desvio Padrão}}{\text{Média}} \times 100\%$


```python
# Medidas de Variabilidade
print("Desvio-padrão =", round(np.std(dados),2))
print("Variância =", round(np.var(dados),2))
print("Coeficiente de Variação (%) =",round(np.std(dados)/np.mean(dados)*100,2))
```

**3 Tabelas de Distribuição de Frequência:** Também conhecida como tabela de frequências, é uma forma organizada de apresentar a contagem ou frequência dos valores individuais ou intervalos de um conjunto de dados. O principal objetivo de uma tabela de distribuição de frequências é resumir e organizar os dados para facilitar a compreensão da distribuição dos valores e suas frequências relativas.

```python
from prettytable import PrettyTable

# Calculando a frequência de cada valor
frequency = {}
for value in dados:
    if value in frequency:
        frequency[value] += 1
    else:
        frequency[value] = 1

# Criando a tabela de distribuição de frequência
tabela = PrettyTable()
tabela.field_names = ["Valor", "Frequência"]

for value, freq in frequency.items():
    tabela.add_row([value, freq])

# Imprimindo a tabela
print(tabela)
```

**4. Gráficos:** Os gráficos são representações visuais de dados que ajudam a comunicar informações de forma clara e compreensível. Eles são amplamente usados em diversas áreas, como estatísticas, ciência, negócios e educação, para apresentar padrões, tendências e relações nos dados.

**4.1 Gráfico de setores:** Um gráfico de setores, também conhecido como gráfico de pizza, é um tipo de visualização que representa a distribuição proporcional de diferentes categorias em um conjunto de dados. Ele é circular e é dividido em "fatias" que correspondem às proporções das categorias representadas. Cada fatia é rotulada com o nome da categoria e, muitas vezes, exibe o percentual que essa categoria representa em relação ao todo.

```python
import matplotlib.pyplot as plt

# Dados
categories = ['A', 'B', 'C', 'D', 'E']
sizes = [25, 15, 20, 10, 30]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'purple']

# Criando o gráfico de setores
plt.pie(sizes, labels=categories, colors=colors, autopct='%1.1f%%', startangle=140)

# Ajustando aspectos
plt.axis('equal')  # Faz com que o gráfico fique circular
plt.title('Gráfico de Setores - Distribuição de Categorias')

# Exibindo o gráfico
plt.show()
```

**4.2 Gráfico de Barras:** É uma representação visual de dados que utiliza barras retangulares para mostrar a relação entre diferentes categorias ou variáveis. Cada barra representa uma categoria específica e sua altura (ou comprimento) é proporcional ao valor ou frequência dessa categoria. Esse tipo de gráfico é usado para comparar quantidades ou distribuições de diferentes categorias de maneira clara e eficaz.

```python
import matplotlib.pyplot as plt

# Preparando dados para o gráfico
values = list(frequency.keys())
frequencies = list(frequency.values())

# Criando o gráfico de barras
plt.bar(values, frequencies, color='blue')

# Adicionando rótulos e título
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.title('Gráfico de Barras - Distribuição de Frequência')

# Exibindo o gráfico
plt.show()
```
